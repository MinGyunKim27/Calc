```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# 데이터 불러오기
data = pd.read_csv('sample_data.csv')
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm
```

```python
data
```

```python
title = data['title']
```

##개행문자 제거

```python
# 'story' 열의 각 행에서 개행문자 제거
data['story'] = data['story'].str.replace('\n', '')
```

```python
# 'story' 열의 각 행에서 개행문자 제거
data['tag'] = data['tag'].str.replace('\n', '')
```

## 스토리 원 닷

```python
data['story'] = data['story'].apply(lambda story_str: np.array([float(num) for num in story_str.strip('[]').split(' ') if num]))


# 'story' 열의 값을 2차원 배열로 변환
story_array = np.vstack(data['story'].values)

# 코사인 유사도 계산
cosine_sim_story = cosine_similarity(story_array, story_array)

# 유사도 행렬 출력
print(cosine_sim_story)
```

```python
# 영화-영화 유저관람 내역에 대한 유사도
movie_similarity_story = pd.DataFrame(cosine_sim_story,
                                index=title, columns=title)
print(movie_similarity_story.shape)
movie_similarity_story.head()
```

##태그 원 닷

```python

# 'tag' 열의 각 행을 벡터로 변환하여 NumPy 배열로 만들기
data['tag'] = data['tag'].apply(lambda story_str: np.array([float(num) for num in story_str.strip('[]').split(' ') if num]))

# 'tag' 열의 값을 2차원 배열로 변환
tag_array = np.vstack(data['tag'].values)

# 코사인 유사도 계산
cosine_sim_tag = cosine_similarity(tag_array, tag_array)

# 유사도 행렬 출력
print(cosine_sim_tag)
```

```python
# 영화-영화 유저관람 내역에 대한 유사도
movie_similarity_tag = pd.DataFrame(cosine_sim_tag,
                                index=title, columns=title)
print(movie_similarity_tag.shape)
movie_similarity_tag.head()
```

##장르 원 닷

```python

# 'genres' 열의 각 행을 벡터로 변환하여 NumPy 배열로 만들기
data['genres'] = data['genres'].apply(lambda genre_list: np.array([int(num) for num in genre_list.strip('[]').split(',')]))

# 'genres' 열의 값을 2차원 배열로 변환
genres_array = np.vstack(data['genres'].values)

# 코사인 유사도 계산
cosine_sim = cosine_similarity(genres_array, genres_array)

# 유사도 행렬 출력
print(cosine_sim)
```

```python
# 영화-영화 유저관람 내역에 대한 유사도
movie_similarity_genre = pd.DataFrame(cosine_sim,
                                index=title, columns=title)
print(movie_similarity_genre.shape)
movie_similarity_genre.head()
```

## 쓰리 닷..?

```python

# 'runtime' 열을 2차원 배열로 만들기
runtime_array = data['runtime'].values.reshape(-1, 1)

# 모든 정보를 합쳐서 2차원 배열 만들기
combined_array = np.hstack([genres_array, runtime_array, tag_array])

# 코사인 유사도 계산
cosine_sim_genre_runtime_tag = cosine_similarity(combined_array, combined_array)

# 유사도 행렬 출력
print(cosine_sim_genre_runtime_tag)
```

```python
# 영화-영화 유저관람 내역에 대한 유사도
movie_similarity_genre_runtime_tag = pd.DataFrame(cosine_sim_genre_runtime_tag,
                                index=title, columns=title)
print(movie_similarity_genre_runtime_tag.shape)
movie_similarity_genre_runtime_tag.head()
```
